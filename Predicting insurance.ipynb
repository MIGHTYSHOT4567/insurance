```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/HP/Desktop/Data/jupyter Folder/Data/new_insurance_data.csv")
df.head(6)
```

```python
df
```

```python
df['Salary'].max()
```

```python
dummies = pd.get_dummies(df.Country)
```

```python
dummies.head(6)
```

```python
merged = pd.concat([df,dummies], axis = "columns")
merged.head(6)
```

```python
final_df = merged.drop(["Country", "USA"], axis = "columns")
final_df.head(6)
```

```python
x = final_df.drop("Insurance Purchase", axis = "columns")
x.head(3)
```

```python
y = df["Insurance Purchase"]
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 30)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
X_test
```

```python
y_test
```

```python
model.predict([[18, 1000, 0,1]])
```

```python
y_predicted = model.predict(X_test)
```

```python
y_predicted
```

```python
model.score(x,y)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_predicted)
print("Accuracy:", accuracy)
```

```python
import joblib as jb
jb.dump(model,"Insurance Model_new")
```
